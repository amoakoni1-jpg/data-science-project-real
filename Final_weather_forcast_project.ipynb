```python
def sigmoid(z):
   
    sig = 1/(np.exp(-z)+1)
    
    return sig
```

```python
def costFunction(X, y, theta=[[0], [0], [0],[0],[0],[0]]):
    m = y.size
    X_extend = np.c_[np.ones(m),X_normalized]
    z = X_extend.dot(theta)
    h = sigmoid(z)
    
    epsilon = 1e-15
    h = np.clip(h, epsilon, 1 - epsilon)
    
    cost = (-1/m)*np.sum(y*np.log(h)+(1-y)*np.log(1-h))
    
    return cost
```

```python
def gradientDescent(X_normalized, y, theta=[[0], [0], [0],[0],[0],[0]], epochs = 1000, learning_rate = 0.0001):
    
    m = y.size
    X_extend = np.c_[np.ones(m),X_normalized]
    j_history = np.zeros(epochs)

    n_print = epochs//10

    for i in range(epochs):
        
        z = X_extend.dot(theta) 
        h = sigmoid(z) 
        epsilon = 1e-15  # Small value to prevent log(0)
        # Clip predictions to avoid 0 and 1
        h = np.clip(h, epsilon, 1 - epsilon)
        
        theta[0] = theta[0]-(learning_rate/m)*np.sum(h-y)
        theta[1] = theta[1]-(learning_rate/m)*np.sum((h-y)*X_normalized[:,0:1])
        theta[2] = theta[2]-(learning_rate/m)*np.sum((h-y)*X_normalized[:,1:2])
        theta[3] = theta[3]-(learning_rate/m)*np.sum((h-y)*X_normalized[:,2:3])
        theta[4] = theta[4]-(learning_rate/m)*np.sum((h-y)*X_normalized[:,3:4])
        theta[5] = theta[5]-(learning_rate/m)*np.sum((h-y)*X_normalized[:,4:5])
        j =  (-1/m) * np.sum(y*np.log(h) + (1-y)*np.log(1-h))
        j_history[i]=j
        
        if (i+1) % n_print == 0:
            print("Iteration", i+1, " | Cost: ", j, "\n")
    
    return  theta, j_history
```

```python


# Fix the other 3 features at their mean values
mean_wind = X[:, 2].mean()
mean_cloud = X[:, 3].mean()
mean_pressure = X[:, 4].mean()

# Adjusted intercept for the 2D plot
adjusted_theta_0 = theta[0] + theta[3]*mean_wind + theta[4]*mean_cloud + theta[5]*mean_pressure

# Plot line using Temperature and Humidity
x_plot = np.array([X[:, 0].min(), X[:, 0].max()])
y_plot = -(1/theta[2]) * (adjusted_theta_0 + theta[1] * x_plot)

plt.plot(X[pos, 0], X[pos, 1], 'b+', label='Rain')
plt.plot(X[neg, 0], X[neg, 1], 'yo', label='Not Rain')
plt.plot(x_plot, y_plot, 'k-', linewidth=3, label='From-Scratch Boundary')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15))
plt.title('Decision Boundary - From Scratch (5 features)')
plt.show()

def predict(X, theta):
    X_extend = np.c_[1,X]
    z = X_extend.dot(theta)
    h = sigmoid(z) 
    if h>=0.5 :
        return 1
    return 0
```
